```python
# filter
import numpy as np
arr = np.array([10,20,3,4,90,200])
arr_f = arr[arr > 20]
print(arr_f)
```

```python
import numpy as np
arr = np.array([10,20,30,6,3,2])
arr_f = arr[arr != 6]
print(arr_f)
```

```python
import numpy as np
arr = np.array([10,20,30,15,3,2,18,8])
arr_f = arr[(arr>10) & (arr<20)]
print(arr_f)
```

```python
# fancy indexing
import numpy as np
arr = np.array([10,20,30,3,4,90,100])
print(arr)
arr_f = arr[[0,2]]
print(arr_f)
```

```python
# where
import numpy as np
arr = np.array([10,20,30,3,4,90,100])
print(arr)
arr_w = np.where(arr > 9 , "True" , "False")
print(arr_w)
```

```python
import numpy as np
arr = np.array([10,20,30,3,4,90,100])

arr_w = np.where((arr>10) & (arr<20) , arr*2 , arr+2)
print(arr_w)
```

```python
# adding data and removing data
import numpy as np
arr1 = np.array([10,20,30])
arr2 = np.array([1,2,3])
arr1_arr2 = np.concatenate((arr1,arr2))
print(arr1_arr2)
```

```python
# concatenate using 2d array
import numpy as np
arr1 = ([[1,2],[3,4]])
arr2 = ([[5,6],[7,8]])
arr1_arr2 = np.concatenate((arr1,arr2),axis=1)
print(arr1_arr2)

```

```python
# vstack

import numpy as np
arr1 = np.array([10,20,30])
arr2 = np.array([4,5,6])

arr_v = np.vstack((arr1,arr2))
print(arr_v)
```

```python
# hstack --> horizontally join column wise
import numpy as np
arr1 = np.array([10,20,30])
arr2 = np.array([4,5,6])

arr_v = np.hstack((arr1,arr2))
print(arr_v)
```

```python
import numpy as np
arr1 = ([[1,2],[3,4]])
arr2 = ([[5,6],[7,8]])
arr1_arr2 = np.vstack((arr1,arr2))
print(arr1_arr2)
```

```python
import numpy as np
arr1 = ([[1,2],[3,4]])
arr2 = ([[5,6],[7,8]])
arr1_arr2 = np.hstack((arr1,arr2))
print(arr1_arr2)
```

```python
import numpy as np
arr = np.array([2,4,6,8,10])
print(np.sum(arr))
print(np.min(arr))
print(np.max(arr))
print(np.mean(arr))
print(np.median(arr))
print(np.std(arr))
print(np.var(arr))
print(np.prod(arr))
print(np.cumsum(arr))
print(np.cumprod(arr))
print(np.argmin(arr))
print(np.argmax(arr))
print(np.abs(arr))
print(np.unique(arr))
print(np.percentile(arr, 50))
print(np.quantile(arr, 0.5))
print(np.ptp(arr))
print(np.any(arr))
print(np.all(arr))



```

```python
#np.sum(a) #-> Sum of all elements
#np.mean(a) #-> Average = Sum of elements / Number of elements
#np.median(a) #-> Middle value after sorting
#np.min(a) #-> Smallest value in array
#np.max(a) #-> Largest value in array
#np.var(a)# mean,difference,square,average | (sum = ddof) -> divide by array length
#np.std(a) #-> Standard Deviation = √Variance -> Measures spread of data
#np.prod(a) #-> Multiplication of all elements
#np.cumsum(a)# -> Cumulative (running) sum
#np.cumprod(a)# -> Cumulative (running) product
#np.argmin(a) #-> Index position of minimum value
#np.argmax(a) #-> Index position of maximum value
#np.abs(a) #-> Converts negative values to positive -> Absolute value (distance from zero)
#np.unique(a)# -> Returns unique values only -> Removes duplicates
#np.percentile(a, 50) #-> Finds percentage-based value -> 50th percentile = Median
#np.quantile(a, 0.5) #3-> Similar to percentile -> 0.5 = 50%
#np.ptp(a) #-> Range = Max - Min
#np.any(a) #-> True if at least one value is True
```
