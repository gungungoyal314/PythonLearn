```python
import numpy as np
import pandas as pd
l = [10,20,30,40]
df = pd.Series(l)
print(df)
```

```python
df = pd.Series(data = l, index =["nikhil","gungun","abhii","hello"])
print(df)
```

```python
d= {'nikhil':10,"devendra":20,"gungun":30,"abhii":40}
df = pd.Series(data=d)
print(df)
```

```python
# pd is nickname of pandas
import pandas as pd
l = [10,20,30,40,50] # list
df = pd.Series(l) # use list in pd
print(df)


# example data
df = pd.Series(data=l)
print(df)

# example index
df = pd.Series(data=l , index=['nikhil','devendra','raj','hello','unknown'])
print(df)

# example key = index | value = value
d = {'nikhil':10,'devendra':20,'raj':30,'hello':40,'unknown':50}
df = pd.Series(data=d)
print(df)

# experiment
```
