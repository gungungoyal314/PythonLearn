```python
# pd is nickname of pandas
import pandas as pd
l = [10,20,30,40,50] # list
df = pd.Series(l) # use list in pd
print(df)


# example data
df = pd.Series(data=l)
print(df)

# example index
df = pd.Series(data=l , index=['nikhil','devendra','raj','hello','unknown'])
print(df)

# example key = index | value = value
d = {'nikhil':10,'devendra':20,'raj':30,'hello':40,'unknown':50}
df = pd.Series(data=d)
print(df)

# experiment
```

```python
#2d
import pandas as pd
data = {
    "name": ["nikhil","gungun","devendra"],
     "age": [10,20,30]
}

df = pd.Series(data)
print(df)


df = pd.DataFrame(data)
print(df)
```

```python
import pandas as pd
data = [["nikhil",10],["devendra",20],["raj",30],["hello",40]]
df = pd.DataFrame(data , columns = ["name",'age'])
print(df)


# add index name
df = pd.DataFrame(data , columns = ["name",'age'],index = ["a","b","c","d"])
print(df)
```
