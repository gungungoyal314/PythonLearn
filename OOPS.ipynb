```python
# oops in python
# Create a class Person with constructor (name, age) and method is_adult().
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def is_adult(self):
    return self.age >= 18


person1 = Person("gungun",20)
person2 = Person("abhishek",16)
print(person1.is_adult())
print(person2.is_adult())
```

```python
class Rectangle:
  def __init__(self , length , width):
    self.length = length
    self.width = width
  def area(self):
      return self.length * self.width


r = Rectangle(10 , 29)
print(r.area())
```

```python
#Student class banao jisme constructor (name, marks) ho aur method has_passed()
 # ho jo 40 ya usse zyada marks par True return kare.


class Student():
  def __init__(self,name,marks):
    self.name = name
    self.marks = marks

  def has_passed(self):
    return self.marks >=40


student1= Student("gungun",39)
student2 =Student("nikhil",41)
Student3 = Student("abhishek",42)


print(student1.has_passed())
print(student2.has_passed())
print(Student3.has_passed())

```

```python
#Circle class banao jisme constructor (radius) ho aur method circumference() ho.

class Circle:
  def __init__(self,radius):
    self.radius = radius

  def circumference(self):
    return 2 * 3.13 * self.radius


C = Circle(10)
print(C.circumference())
```

```python
# Constructor (owner, balance) ho. Methods:

# deposit(amount)
# withdraw(amount)


class BankAccount:
  def __init__(self , owner , balance):
    self.owner = owner
    self.balance = balance


  def deposit(self , amount):
    self.amount =+1

  def withdraw(self,amount):
    if amount <= self.balance:
      self.balance -= amount
    else:
      print("insufficient balance")




acc = BankAccount("gungun",1000)
acc.deposit(500)
acc.withdraw(200)
print(acc.balance)

```

```python
#Constructor (brand, speed) ho aur method accelerate() speed ko 10 se increase kare.


class Car:
  def __init__(self,brand,speed):
    self.brand = brand
    self.speed = speed

  def accelerate(self):
    self.speed =+10


car1 = Car("audi",100)
car2 = Car("bmw",200)

car1.accelerate()
car2.accelerate()
print(car1.speed)
print(car2.speed)
```

```python
# Constructor (name, salary) ho aur method annual_salary() yearly salary return kare.
class Employee:
  def __init__(self,name,salary):
    self.name = name
    self.salary = salary

  def annual_salary(self):
    return self.salary * 12


emp1 = Employee("gungun",10000)
emp2 = Employee("abhishek",20000)
emp1.annual_salary()
emp2.annual_salary()
```
