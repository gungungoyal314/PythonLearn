```python
# inheritence in python
class Animal:
    def sound(self):
        print("Animals make sounds")

class Dog(Animal):
    def bark(self):
        print("Dog barks")

class Cat(Animal):
  def meow(self):
    print("Cat Meows")

d = Dog()
c = Cat()
c.meow()
d.sound()   # Parent class method
d.bark()    # Child class method
```

```python
class Grandparent:
  def show(self):
    print("Grandparent")


class Parent(Grandparent):
  def show1(self):
    print("Parent")


class Child(Parent):
  def show2(Self):
    print("Child")


c = Child()
c.show()
c.show1()
c.show2()
```

```python
class Car:
  def model(self):
    print("BMW")


class BMW(Car):
  def show1(self):
    print("AUDI")

c = Car()
c.model()
b = BMW()
b.show1()
```
