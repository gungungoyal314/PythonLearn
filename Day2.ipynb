```python
# Bitwise operator work on binary digits
num1 = 10
num2 = 20
print(num1 & num2)
print(num1 | num2)
print(num1 ^ num2)
print(~num1)
print(~num2)
print(num1 << 2)
print(num2 >> 2)
```
