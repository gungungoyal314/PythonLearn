```python
# handle missing values
import pandas as pd
import numpy as np
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":["Hello","Gungun","Abhii","Priyanshu","Great"]
}
df = pd.DataFrame(data)
print(df)

# null values filter
df.isnull()

# sum of null
df.isnull().sum()

# dropna  --> drop complete row
df.dropna()
print(df)


# fillna
df.fillna(5)
```

```python
# matplotlib
import matplotlib.pyplot as plt
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500],
}
x = data['experience']
y = data['salary']
# plt.plot --> line chart
plt.plot(data["experience"],data["salary"],marker = '*',markersize = 14,color = 'pink')
plt.figure(figsize=(6,4))
plt.savefig("hello.jpeg")
plt.show()
```

```python
# pie chart
x = ['apple','banana','carrot','mango']
y = [30,50,70,5]

plt.figure(figsize=(4,4))
plt.pie(y,labels = x,autopct = "%1.1f%%",colors=['red','yellow','orange','green'])
plt.show()
```

```python
# subplots
import matplotlib.pyplot as plt #
# pie chart data
x = ["apple","banana","carrot","mango"]
y = [30,50,70,5]
fig , aux = plt.subplots(1,2)
aux[0].pie(y,labels=x)

# line chart data
data = {
    "experience":[1,2,3,4,5],
    "salary":[100,200,300,400,500]
}
x = data['experience']
y = data['salary']

aux[1].plot(x,y)
plt.tight_layout()
plt.show()
```
