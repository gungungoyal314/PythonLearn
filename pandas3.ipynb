```python
# handle missing values
import pandas as pd
import numpy as np
data = {
    "experience":[1,np.nan,3,4,5],
    "salary":[100,200,300,400,np.nan],
    "name":["Hello","Gungun","Abhii","Priyanshu","Great"]
}
df = pd.DataFrame(data)
print(df)

# null values filter
df.isnull()

# sum of null
df.isnull().sum()

# dropna  --> drop complete row
df.dropna()
print(df)


# fillna
df.fillna(5)
```
