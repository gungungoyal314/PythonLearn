```python
import numpy as np
arr = np.array([10,35,40,7,15,20])
print(arr)
print(np.sort(arr))   # ascending
print(np.sort(arr)[::-1]) # descending


```

```python
import numpy as np

arr1 = arr.reshape(2,3)

print(arr1)
print(np.sort(arr1))   # ascending
print(np.sort(arr1)[::-1]) # descending
```

```python
# filter
import numpy as np
arr = np.array([10,20,30,40])
print(arr[arr> 30])
print(arr[0])
print(arr[0:3])
print(arr[0]+arr[1])
```
