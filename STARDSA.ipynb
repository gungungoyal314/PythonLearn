```python
for i in range(4):
    for j in range(4):
        print("*", end=" ")
    print()
```

```python
for i in range(1, 6):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
for i in range(5, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```

```python
rows = 5

for i in range(rows):
    print(" " * (rows - i - 1), end="")
    print("*" * (2 * i + 1))
```

```python
rows = 5

for i in range(rows):
    print(" " * i, end="")
    print("*" * (2 * (rows - i) - 1))
```

```python
rows = 5

for i in range(rows):
    print(" " * (rows - i - 1), end="")
    print("*" * (2 * i + 1))

for i in range(rows - 2, -1, -1):
    print(" " * (rows - i - 1), end="")
    print("*" * (2 * i + 1))
```

```python
n = 5

for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end="")
        else:
            print(" ", end="")
    print()
```
