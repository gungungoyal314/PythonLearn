```python
numbers = [1, 2, 3, 4, 5, 6]

even_numbers = [x for x in numbers if x % 2 == 0]

print(even_numbers)
```

```python
words = ["apple", "banana", "cherry"]

uppercase_words = [word.upper() for word in words]

print(uppercase_words)
```

```python
words = ["APPLE", "BANANA", "CHERRY"]

lowercase_words = [word.lower() for word in words]

print(lowercase_words)
```
